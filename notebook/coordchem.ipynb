## 📘 Project Overview

Coordination chemistry plays a central role in inorganic chemistry and appears in catalysis, bioinorganic chemistry, medicinal chemistry, supramolecular chemistry and materials science.

However, the study of coordination compounds often requires repetitive analytical tasks:

- oxidation state determination,
- ligand identification,
- electron counting,
- geometry prediction,
- nomenclature generation,
- and molecular visualization.

To simplify these operations, we developed **CoordChemPy**, an educational Python package dedicated to coordination chemistry analysis.

The objective of the project was to create a toolkit that remains:

- scientifically rigorous,
- intuitive for undergraduate students,
- easy to use,
- and independent from advanced cheminformatics knowledge.

Unlike many chemistry packages, CoordChemPy does not require:

- SMILES notation,
- graph theory knowledge,
- external molecular databases,
- or complicated molecular file formats.

Users can directly manipulate coordination compounds using a simplified inorganic chemistry notation.

---

# 🎯 Motivation

The primary goal of CoordChemPy is to bridge the gap between inorganic chemistry theory and computational chemistry tools.

Many existing cheminformatics libraries are powerful but difficult to approach for students beginning coordination chemistry.

CoordChemPy was therefore designed with three priorities:

1. Accessibility
2. Educational value
3. Scientific relevance

The package allows students to rapidly explore coordination compounds while reinforcing theoretical inorganic chemistry concepts.

---

# ⚠️ Package Limitations

Although CoordChemPy implements numerous coordination chemistry features, several simplifications were necessary during development.

Currently, the package supports:

- classical transition metals,
- mononuclear complexes,
- symmetric dinuclear complexes,
- common ligand environments,
- standard coordination geometries.

The current limitations include:

- non exhaustive databases (ligand and counter ions),
- no lanthanides or actinides,
- no heterobimetallic systems,
- no complexes containing more than two metal centers,
- empirical stability estimations,
- simplified geometry prediction algorithms,
- approximate stereoisomer counting.

Despite these limitations, CoordChemPy already provides a robust educational framework for coordination chemistry exploration.

# 🧩 Coordination Complex Syntax

CoordChemPy uses a simplified notation inspired by classical inorganic chemistry conventions.

---

## General Rules

- Coordination spheres are written inside square brackets `[]`
- Metals appear first inside the brackets
- The prefixe before the metal symbol indicates the metal-metal bond nature
- Ligands are written inside parentheses `()`
- Stoichiometric coefficients appear after ligands
- Charges are written outside brackets
- Bridging ligands use the prefix `m-`

---

## Metal–Metal Bond Prefixes

| Prefix | Meaning |
|---|---|
| s | single bond |
| d | double bond |
| t | triple bond |
| q | quadruple bond |

---

## Examples of valid formulas ✅

| Compound Type | Example |
|---|---|
| Neutral complex | `[Pt(Cl)2(NH3)2]` |
| Charged complex | `[Fe(CN)6]3-` |
| Dinuclear complex | `[sCo2(m-OH)2(NH3)8]4+` |
| Complex with counter ions | `[Cr(en)3]Cl3` |

---

This notation allows rapid manipulation of coordination compounds without requiring advanced computational chemistry knowledge.

# 🔍 Formula Verification and Parsing

CoordChemPy first validates the coordination compounds input format using regular expressions before extracting:

- metal centers,
- ligands,
- stoichiometric coefficients,
- and charges.

This validation step simplifies all subsequent chemical analysis.

## `cc.validate_formula()`

Before deep analysis, this function ensures the whole compound (coordination sphere + counter-ions) follows all implemented chemical rules.

````python
formula = "[Cr(H2O)6]3+"
counter_ions = "(Cl)3"

cc.validate_formula(formula, counter_ions)
````

## `cc.parse_ligands()`

````python
formula = "[Cu2(OAc)4(H2O)2]"

print(cc.parse_ligands(formula))
````

## `cc.parse_metal()`

````python
print(cc.parse_metal(formula))
````

## `cc.parse_counter_ions()`

````python
counter_ions = "(NO3)2"

print(cc.parse_counter_ions(counter_ions))
````

The output of these parsing functions is designed to be easily reused in other dependent functions throughout the package. For metal parsing, a simple list is sufficient, since the package only supports a single metal center. For ligands and counter-ions, the functions return three lists:
- All elements individually
- Each element appearing once, in a defined order
- The corresponding coefficient for each element in that same order  

This structure allows users to select the format that best fits their needs depending on the context.

# 🧪 Global Compound Analysis

The function `cc.analyse_compound()` gathers several coordination chemistry descriptors simultaneously. Here is a list just name a few:

- clean formula rendering
- IUPAC-like naming
- Oxidation state, electronic structure and charge of the metal
- geometry prediction
- number of stereoisomer
- stability estimation
- Jahn-Teller distortion intensity ...

This function represents the core analysis module of CoordChemPy. Here is an exemple with the cisplatin compound:

````python
formula = "[Pt(Cl)2(NH3)2]"

cc.analyse_compound(formula)
````

# ✨ Formula Rendering

The function `cc.get_clean_formula()` converts raw user inputs into LaTeX-formatted chemical formulas suitable for publication-quality rendering.

````python
from IPython.display import Markdown, display

formula = "[Cr(H2O)6]3+"
counter_ions = "(Cl)3"

display(Markdown(cc.get_clean_formula(formula, counter_ions)))
````

# 🏷️ IUPAC Naming

CoordChemPy automatically generates systematic coordination compound names following IUPAC nomenclature rules. The naming of a compound is limited by the ligand database included in the package. However, this database covers the most common ligands and counter-ions, ensuring that the naming_compound() function remains applicable to a wide range of standard coordination complexes.

````python
formula = "[Cr(H2O)6]3+"
counter_ions = "(Cl)3"

cc.naming_compound(formula, counter_ions)
````

In all function that can take as an argument counter_ions, the latter is optional.

# ⚡ Oxidation State Determination

These informations are calculated from:

- ligand charges,
- counter ions,
- and total complex charge.

````python
formula = "[Fe(CN)6]3-"

print(f"d{cc.oxidation_state(formula)[0]}")
````

# 🧮 Electron Counting

Electron counting is one of the most important concepts in coordination chemistry.

CoordChemPy calculate the valence electron count using ligand donor properties and oxidation states.

````python
formula = "[Ni(CO)4]"

cc.electron_count(formula)
````

# Number of isomers

Using a theoretical table, the package can determine whether a compound has isomers. The function is based on the number of different ligands around the metal center. It returns a tuple indicating, first, the number of stereoisomers and, second, the number of enantiomeric pairs.

````python
formula = "[Fe(CO)3(H2O)2(Cl)]"

cc.isomers(formula)
````

# 📐 Geometry Prediction

CoordChemPy predicts the most probable geometry according to:

- coordination number,
- electronic configuration

````python
formula = "[Ni(CN)4]2-"

cc.get_geometry(formula)[1]
````

# 🌈 Crystal Field, Jahn–Teller and spin

Certain coordination compounds exhibit structural distortions linked to electronic degeneracies.

CoordChemPy includes simplified crystal field considerations and can identify probable Jahn–Teller distortions for selected electronic configurations. The counter ions are not taken into account in these calculation

This feature is especially useful for educational demonstrations in inorganic chemistry courses. Thanks to the Jahn-Teller analysis, one can know for octahedral if there is distortion or not giving more information about the geometry of the complex.

````python
formula = "[Cu(H2O)6]2+"

print(cc.jahn_teller_distortion(formula))
print(cc.determine_spin_state(formula)[0])
````

# 💯 Stability test and duel

The Stability Duel feature allows you to quantitatively compare the stability of two coordination complexes side by side using the scoring algorithm implemented in the package. The resulting score has no strict scientific meaning; it is intended only as a relative scale for comparison within the framework of the model.

````python
compound_A = "[Fe(CO)6]"
compound_B = "[Cr(NH3)6]"

cc.stability_duel(compound_A, compound_B)
````

While it provides a quick theoretical benchmark, coordination chemistry is highly complex. This score can lack precision or generate errors depending on specific ligand geometries, steric hindrance, or unusual oxidation states. It should be used as a general guide rather than an absolute experimental truth.

# 💾 Exporting Structures

CoordChemPy constructs an internal ASE (Atomic Simulation Environment) `Atoms` object that can easily be exported to standard computational chemistry formats such as XYZ and CIF.

````python
formula = "[Pt(Cl)2(NH3)2]"

# Create the underlying ASE compound object
compound = cc.create_compound_render(formula)
print(f"ASE Object created with {len(compound)} atoms.")

# Export files locally
xyz_path = cc.export_xyz(formula, "assets/complex.xyz")
cif_path = cc.export_cif(formula, "assets/complex.cif")

print(f"Files successfully exported to: {xyz_path} and {cif_path}")
````

# 📊 Comparative Analysis of Multiple Complexes

One important advantage of computational chemistry tools is the ability to rapidly compare multiple compounds.

````python
complexes = [
    "[Fe(CN)6]3-",
    "[Ni(CO)4]",
    "[Pt(Cl)2(NH3)2]",
    "[Cu(H2O)6]2+"
]

for compound in complexes:
    print("="*60)
    print(compound)
    cc.analyse_compound(compound)
    print("\n")
````
