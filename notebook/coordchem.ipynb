## 📘 About this Notebook

Coordination chemistry is a central field of inorganic chemistry, with applications ranging from catalysis and bioinorganic chemistry to materials science and medicinal chemistry. - automatic coordination complex parsing,
- oxidation state determination,
- electron counting,
- geometry prediction,
- stereoisomer estimation,

- and interactive 3D rendering.

This notebook presents the main functionalities of the package through practical examples and interactive demonstrations.

## Introduction


# 🎯 Motivation

Coordination chemistry is often perceived as difficult by undergraduate students because of the large number of conventions, nomenclature rules and structural possibilities involved.

The main objective of CoordChemPy is therefore to provide a simple and intuitive toolkit capable of automating many classical inorganic chemistry tasks while remaining accessible to users with limited programming experience.

Unlike many cheminformatics packages, CoordChemPy does not require:
- knowledge of SMILES notation,
- advanced graph theory,
- external databases,
- or complicated molecular input formats.

Instead, users can directly input coordination compounds using an intuitive notation inspired by classical inorganic chemistry writing conventions.

The package was specifically designed for:
- undergraduate inorganic chemistry students,
- teaching assistants,
- educators,
- and researchers needing rapid coordination compound analysis.

CoordChemPy aims to bridge the gap between inorganic chemistry theory and computational chemistry tools.

# ⚠️ Package Limitations

Although CoordChemPy provides many functionalities for coordination chemistry analysis, some approximations and restrictions were necessary during development.

Currently, the package supports:

- Classical transition metals only
- Mononuclear or symmetric dinuclear complexes
- Standard coordination geometries
- A curated ligand and counter-ion database

The current limitations are the following:

- The ligand and counter-ion databases are not exhaustive
- Lanthanides, actinides and superheavy elements are not implemented
- Heterobimetallic complexes are not supported
- Polynuclear complexes containing more than two metal centers are excluded
- Some geometry predictions rely on simplified heuristic models
- Stability predictions are empirical and should not replace experimental measurements
- Stereoisomer counting is limited for highly complex coordination environments
- Crystal field effects are approximated
- Dynamic ligand behavior is not considered

Despite these limitations, CoordChemPy already provides a robust educational and exploratory framework for coordination chemistry.

# 🧩 Coordination Complex Input Syntax

CoordChemPy uses a simplified textual representation for coordination compounds.

The notation follows a set of strict but intuitive rules:

---

## General Rules

- The coordination sphere must be written inside square brackets `[]`
- Metals always appear first
- Ligands are always written inside parentheses `()`
- Well-known abbreviation of ligands can be used and need to be written in lowercase
- Stoichiometric coefficients are written after ligands
- Charges are written outside brackets
- Bridging ligands are prefixed using `m-`

---

## Examples of valid formulas ✅

| Compound Type | Example |
|---|---|
| Neutral complex | `[Pt(ox)(NH3)2]` |
| Charged complex | `[Fe(CN)6]3-` |
| Dinuclear complex | `[sCo2(m-OH)2(NH3)8]4+` |
| Complex with counter ions | `[Cr(H2O)6]Cl3` |

---

## Supported Bond Prefixes

| Prefix | Meaning |
|---|---|
| s | single metal-metal bond |
| d | double bond |
| t | triple bond |
| q | quadruple bond |

---

This simplified syntax allows rapid manipulation of coordination compounds without requiring advanced cheminformatics knowledge. In the case of an error the databases are easy to access and well structured which allows quick search of ligands or counter ions for instance.

# 🔍 Formula Verification and Parsing

CoordChemPy first validates the formula syntax using regular expressions before extracting:
- metals,
- ligands,
- stoichiometric coefficients,
- and charges.

This two-step parsing strategy greatly simplifies downstream analysis.

## `cc.formula_verif_and_parsing()`

This function verifies the global syntax of the coordination complex.

````python
formula = "[Cu2(OAc)4(H2O)2]"

cc.formula_verif_and_parsing(formula)
````

If a re.Match object is created, it means taht the formula has the correct sythax

## `cc.parse_ligands()`

````python
formula = "[Cu2(OAc)4(H2O)2]"

print(cc.parse_ligands(formula))
````

## `cc.parse_metal()`

````python
print(cc.parse_metal(formula))
````

## `cc.parse_counter_ions()`

````python
counter_ions = "(NO3)2"

print(cc.parse_counter_ions(counter_ions))
````

# 🧪 Global Compound Analysis

The function `cc.analyse_compound()` gathers several important coordination chemistry descriptors simultaneously:

- Formula cleaning
- IUPAC naming
- Oxidation state determination
- Electron counting
- Electronic configuration
- Geometry prediction
- Isomer estimation
- Stability scoring

This function serves as the central high-level analysis tool of the package.

````python
formula = "[Pt(Cl)2(NH3)2]"

cc.analyse_compound(formula)
````

# ✨ Formula Rendering

The function `cc.get_clean_formula()` converts raw input strings into LaTeX-formatted chemical formulas suitable for publication-quality rendering.

````python
from IPython.display import Markdown

counter_ions = "(Cl)3"
formula = "[Cr(H2O)6]3+"

Markdown(cc.get_clean_formula(formula, counter_ions))
````

# 🏷️ IUPAC Naming

CoordChemPy automatically generates systematic coordination compound names inspired by IUPAC nomenclature conventions.

````python
formula = "[Cr(H2O)6]3+"
counter_ions = "(Cl)3"

cc.naming_compound(formula, counter_ions)
````

# ⚡ Oxidation State Determination

Oxidation states are automatically determined from:
- ligand charges,
- counter ions,
- and overall coordination sphere charge.

````python
formula = "[Fe(CN)6]3-"

cc.oxidation_state(formula)
````

# 📐 Geometry Prediction

CoordChemPy predicts the most probable coordination geometry based on:
- coordination number,
- ligand environment,
- and electron configuration.

````python
formula = "[Ni(CN)4]2-"

cc.analyse_compound(formula)
````

# 🌐 3D Structure Visualization

CoordChemPy includes an interactive molecular visualization engine based on py3Dmol.

This allows rapid exploration of coordination geometries directly inside Jupyter notebooks.

````python
formula = "[Co(NH3)2(CO)4]"

cc.render_complex(formula)
````
