## 📘 Project Overview

Coordination chemistry plays a central role in inorganic chemistry and appears in catalysis, bioinorganic chemistry, medicinal chemistry, supramolecular chemistry and materials science.

However, the study of coordination compounds often requires repetitive analytical tasks:

- oxidation state determination,
- ligand identification,
- electron counting,
- geometry prediction,
- nomenclature generation,
- and molecular visualization.

To simplify these operations, we developed **CoordChemPy**, an educational Python package dedicated to coordination chemistry analysis.

The objective of the project was to create a toolkit that remains:

- scientifically rigorous,
- intuitive for undergraduate students,
- easy to use,
- and independent from advanced cheminformatics knowledge.

Unlike many chemistry packages, CoordChemPy does not require:

- SMILES notation,
- graph theory knowledge,
- external molecular databases,
- or complicated molecular file formats.

Users can directly manipulate coordination compounds using a simplified inorganic chemistry notation.

---

# 🎯 Motivation

The primary goal of CoordChemPy is to bridge the gap between inorganic chemistry theory and computational chemistry tools.

Many existing cheminformatics libraries are powerful but difficult to approach for students beginning coordination chemistry.

CoordChemPy was therefore designed with three priorities:

1. Accessibility
2. Educational value
3. Scientific relevance

The package allows students to rapidly explore coordination compounds while reinforcing theoretical inorganic chemistry concepts.

---

# ⚠️ Package Limitations

Although CoordChemPy implements numerous coordination chemistry features, several simplifications were necessary during development.

Currently, the package supports:

- classical transition metals,
- mononuclear complexes,
- symmetric dinuclear complexes,
- common ligand environments,
- and standard coordination geometries.

The current limitations include:

- non exhaustive ligand databases,
- no lanthanides or actinides,
- no heterobimetallic systems,
- no complexes containing more than two metal centers,
- empirical stability estimations,
- simplified geometry prediction algorithms,
- and approximate stereoisomer counting.

Despite these limitations, CoordChemPy already provides a robust educational framework for coordination chemistry exploration.

# 🧩 Coordination Complex Syntax

CoordChemPy uses a simplified notation inspired by classical inorganic chemistry conventions.

---

## General Rules

- Coordination spheres are written inside square brackets `[]`
- Metals appear first
- Ligands are written inside parentheses `()`
- Stoichiometric coefficients appear after ligands
- Charges are written outside brackets
- Bridging ligands use the prefix `m-`

---

## Metal–Metal Bond Prefixes

| Prefix | Meaning |
|---|---|
| s | single bond |
| d | double bond |
| t | triple bond |
| q | quadruple bond |

---

## Examples

| Compound Type | Example |
|---|---|
| Neutral complex | `[Pt(Cl)2(NH3)2]` |
| Charged complex | `[Fe(CN)6]3-` |
| Dinuclear complex | `[sCo2(m-OH)2(NH3)8]4+` |
| Complex with counter ions | `[Cr(H2O)6]Cl3` |

---

This notation allows rapid manipulation of coordination compounds without requiring advanced computational chemistry knowledge.

# 🔍 Formula Verification and Parsing

CoordChemPy first validates coordination compounds using regular expressions before extracting:

- metal centers,
- ligands,
- stoichiometric coefficients,
- and charges.

This validation step simplifies all subsequent chemical analysis.

## `cc.formula_verif_and_parsing()`

````python
formula = "[Cu2(OAc)4(H2O)2]"

cc.formula_verif_and_parsing(formula)
````

## `cc.parse_ligands()`

````python
formula = "[Cu2(OAc)4(H2O)2]"

print(cc.parse_ligands(formula))
````

## `cc.parse_metal()`

````python
print(cc.parse_metal(formula))
````

## `cc.parse_counter_ions()`

````python
counter_ions = "(NO3)2"

print(cc.parse_counter_ions(counter_ions))
````

## `cc.validate_formula()`

Before deep analysis, this function ensures the whole compound (coordination sphere + counter-ions) follows all implemented chemical rules.

````python
formula = "[Cr(H2O)6]3+"
counter_ions = "(Cl)3"

cc.validate_formula(formula, counter_ions)
````

## `cc.atom_symbols()`

Extracts the complete and ordered list of atomic symbols composing the coordination compound. This step is crucial for accurate 3D geometry generation.

````python
formula = "[Pt(Cl)2(NH3)2]"

print(cc.atom_symbols(formula))
````

# 🧪 Global Compound Analysis

The function `cc.analyse_compound()` gathers several coordination chemistry descriptors simultaneously:

- clean formula rendering,
- IUPAC-like naming,
- oxidation state determination,
- electronic configuration,
- electron counting,
- geometry prediction,
- stereoisomer estimation,
- and stability scoring.

This function represents the core analysis module of CoordChemPy.

````python
formula = "[Pt(Cl)2(NH3)2]"

cc.analyse_compound(formula)
````

# ✨ Formula Rendering

The function `cc.get_clean_formula()` converts raw user inputs into LaTeX-formatted chemical formulas suitable for publication-quality rendering.

````python
formula = "[Cr(H2O)6]3+"
counter_ions = "(Cl)3"

display(Markdown(cc.get_clean_formula(formula, counter_ions)))
````

# 🖋️ Text Formatting Utilities

CoordChemPy provides safe formatters for converting raw strings into cleaner Unicode-rendered formulas and standardizing charge displays for terminal or text outputs.

````python
print("Beautified formula: ", cc.beautify_formula("H2O"))
print("Beautified formula: ", cc.beautify_formula("NH3"))
print("Formatted charge (+3): ", cc.display_charge(3))
print("Formatted charge (-2): ", cc.display_charge(-2))
````

# 🏷️ IUPAC Naming

CoordChemPy automatically generates systematic coordination compound names inspired by IUPAC nomenclature rules.

````python
formula = "[Cr(H2O)6]3+"
counter_ions = "(Cl)3"

cc.naming_compound(formula, counter_ions)
````

# ⚡ Oxidation State Determination

The oxidation state is automatically calculated from:

- ligand charges,
- counter ions,
- and total complex charge.

````python
formula = "[Fe(CN)6]3-"

cc.oxidation_state(formula)
````

# 🧮 Electron Counting

Electron counting is one of the most important concepts in coordination chemistry.

CoordChemPy estimates the valence electron count using ligand donor properties and oxidation states.

````python
formula = "[Ni(CO)4]"

cc.analyse_compound(formula)
````

# 📐 Geometry Prediction

CoordChemPy predicts the most probable geometry according to:

- coordination number,
- ligand field considerations,
- electronic configuration,
- and empirical coordination chemistry rules.

````python
formula = "[Ni(CN)4]2-"

cc.analyse_compound(formula)
````

# 🌈 Crystal Field and Jahn–Teller Discussion

Certain coordination compounds exhibit structural distortions linked to electronic degeneracies.

CoordChemPy includes simplified crystal field considerations and can identify probable Jahn–Teller distortions for selected electronic configurations.

This feature is especially useful for educational demonstrations in inorganic chemistry courses.

````python
formula = "[Cu(H2O)6]2+"

cc.analyse_compound(formula)
````

# 🌐 3D Molecular Visualization

CoordChemPy integrates interactive 3D visualization using py3Dmol.

This allows direct visualization of coordination geometries inside Jupyter notebooks.

````python
formula = "[Co(NH3)2(CO)4]"

cc.render_complex(formula)
````

# 🚀 Full Pipeline: Analyse & Render

The function `cc.analyse_and_render()` executes a complete coordination chemistry analysis and generates the corresponding 3D visualization simultaneously.

````python
formula = "[Fe(CN)6]3-"

analysis, view = cc.analyse_and_render(formula)

display(Markdown(analysis))
view
````

# 💾 Exporting Structures

CoordChemPy constructs an internal ASE (Atomic Simulation Environment) `Atoms` object that can easily be exported to standard computational chemistry formats such as XYZ and CIF.

````python
formula = "[Pt(Cl)2(NH3)2]"

# Create the underlying ASE compound object
compound = cc.create_compound_render(formula)
print(f"ASE Object created with {len(compound)} atoms.")

# Export files locally
xyz_path = cc.export_xyz(formula, "complex.xyz")
cif_path = cc.export_cif(formula, "complex.cif")

print(f"Files successfully exported to: {xyz_path} and {cif_path}")
````

# 📊 Comparative Analysis of Multiple Complexes

One important advantage of computational chemistry tools is the ability to rapidly compare multiple compounds.

````python
complexes = [
    "[Fe(CN)6]3-",
    "[Ni(CO)4]",
    "[Pt(Cl)2(NH3)2]",
    "[Cu(H2O)6]2+"
]

for compound in complexes:
    print("="*60)
    print(compound)
    cc.analyse_compound(compound)
    print("\n")
````
