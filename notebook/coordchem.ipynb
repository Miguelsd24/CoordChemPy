## Package Limitation
In order to generate a fully functional python package, some approximations were made and some functions are not implemented, at least for now.
Thus, the latter does not cover all coordination complexes possibilities, here is a list of what is included in our CoordChemPy:
* The ligands and counter ions database is not exhaustive
* Only classical trasition metals were considered. Lanthanides, actinides and heavy synthetic metals (Rf --> Cn) were excluded
* Coordination complexes with more than two metal centers or two different metal centers are not incorporated. Also, if the complex has two metal centers, the complex must be symmetric with respect to the metals
* ...

### Complexe analysis
The first part of our package focuses on the analysis of coordination complexes. This package has various parsing functions which restrict the format input of coordination complexes, however the rules are simple to allow anyone to use CoordChemPy.
* The coordination sphere must be written indicated within square brackets
* Inside the brackets, the metal come first, with its stoechiometric coefficient right after. As mentionned before this package contrains this coefficient to 1 or 2, any other values will lead to errors.
* The ligands come after the metals and are always indicated inside parenthesis even if it is a single atom, the stoechiometric coefficient comes outside the parenthesis after the ligand
* Bridging ligand are announced by a prefix "m-" which lies inside the parethesis  before the ligand formula
* If there is two metals, a single, double, triple or quadruple bond can be added by adding the letter s,d,t,q respectively, before the metal symbol. No letter corresponds to no bond.
* The coordination sphere charge must be indicated outside the brackets at the end of the formula. Nothing corresponds to a neutral charge.
  
Our package also takes into account coordination complexes with their counter ions. One must define another variable for the package to take into account the counter ions. Also, the counter ions formula follow the same rules as statted previously
  
For instance, some valid formulas are :

````python
counter_ions = "(K)3"
counter_ions = "(NO3)2"

formula = "[V(Cl)4]"
formula = "[Ti(H2O)5(CO)2]2+"
formula = "[sCo2(m-NH2)(m-OH)(NH3)8]4+"
````

#### Verification and parsing functions


##### cc.formula_verif_and_parsing()  
First, using regular expression which enable one to find pattern in a given string, the formula format was verified. This step does not involve any database matching, it is just the carhacters that are verified to be in a correct spot in a correct order. This first verification makes the parsing of the metals, ligands and coefficient easier moving forward. The regex used in this function allows a first parsing which seperates the metal, ligands and charge of the complex.
##### cc.parse_metal(), cc.parse_ligands(), cc.parse_counter_ions()
These function verify that the input match the local database of the package (see the .json files inside the data folder) and for the parse.ligand() function a second parsing happens to seperate the ligands and their respective coefficients.  
Here is an exemple of what these functions return. Note that the output format of the latter depedns on how the information is processed along the package: parse_metal() returns a list as for parse_ligands and counter_ions they return tuples of lists.

````python
counter_ions = "(NO3)2"
formula = "[Cu2(OAc)4(H2O)2]"

cc.formula_verif_and_parsing(formula)

print(cc.parse_counter_ions(counter_ions))
print(cc.parse_ligands(formula))
print(cc.parse_metal(formula))
````

#### cc.analyse_compound()
The first feature that our package implements is the analysis of a coordination compound. The function displays the following informations:
* A clean version of the input formula, 
* The IUPAC name of the compound, 
* The metal center oxidation state along with its electronic structure, 
* The number of valance electrons, 
* The number of isomers, 
* A relative stability score  
* The propbable geometry. 
Here's an exemple with the cisplatin compound:

````python
formula = "[Pt(Cl)2(NH3)2]" # Cisplatin

cc.analyse_compound(formula)
````

#### cc.get_clean_formula()
We will now go over each function used to get all the information about a compound. The function get_clean_formula(formula, counter_ions) takes two arguments which are respectively the coordination complex and the counter ions formulas. This is the case for the majority of the functions. The function returns the formula following IUPAC rules, in particular, in a coordination complex the cation is written before the anion. Also, the output format is LaTeX which allows a good display of the subscripts and superscripts in Markdown.  
If we use the function separetely we have to explicitely set the Markdown render. By opposition, in analyse_compound(), the Markdown render is already integrated in the function.  
Here is an exemple, with a chromium complex and chloride counter ions:

````python
counter_ions = "(Cl)3"
formula = "[Cr(H2O)6]"

Markdown(cc.get_clean_formula(formula, counter_ions))
````

If counter_ions is not defined, the function runs normally as it only considers the cooordination complex. All package's functions are written to run even if no counter ions formula is defined because the functions analyse focuses on the coordination complex.

#### cc.naming_compound()
The function naming_compound(formula, counter_ions) takes two arguments and return the IUPAC name of the input coordination compound. Therefore the cation (counter ions or coordination sphere) come first followed by the anion. Ligands are arranged in alphabetical order and two sets of prefixes are used according to IUPAC rules.
One can get more information at the International Union of Pure and Applied Chemistry site: iupac.org  
Here is an exemple, with the same chromium compound as before:

````python
cc.naming_compound(formula, counter_ions)
````
